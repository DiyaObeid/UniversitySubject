```python
from datasets import load_dataset

# Load the IMDb dataset
dataset = load_dataset("imdb")

# Check the structure of the dataset
print(dataset)
```

```python
from transformers import AutoTokenizer

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Define the preprocessing function
def preprocess_function(examples):
    return tokenizer(examples['text'], truncation=True, padding='max_length', max_length=128)

# Apply the preprocessing to the dataset
tokenized_datasets = dataset.map(preprocess_function, batched=True)
```

```python
from transformers import AutoModelForSequenceClassification

# Load the BERT model with a sequence classification head
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)

```

```python
# to mimize un used varible in  memmory and cheack if GPU is avalible
import torch
torch.cuda.is_available()
# import torch
# print(torch.cuda.memory_summary(device=None, abbreviated=False))
# import gc
# torch.cuda.empty_cache()
# gc.collect()
```

```python
from transformers import TrainingArguments, Trainer

# Define the training arguments
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=4,  # Reduce the batch size
    per_device_eval_batch_size=4,   # Reduce the batch size
    num_train_epochs=3,
    weight_decay=0.01,
)


# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['test'],
)
```

```python
# Train the model
trainer.train()
```

```python
# Evaluate the model
metrics = trainer.evaluate()

print(metrics)
```

```python
# Save the model
model.save_pretrained("/content/drive/MyDrive/text_classification_model")
```

```python
import torch

# Check if GPU is available and set the device accordingly
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Move the model to the appropriate device
model.to(device)

# Example text for classification
# texts = ["I hate this movie!", "This movie was boring..."]
texts = ["I adore this movie when i watch it"]


# Preprocess the text
inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

# Remove 'token_type_ids' from inputs if it exists
inputs.pop("token_type_ids", None)

# Move the inputs to the same device as the model
inputs = {key: value.to(device) for key, value in inputs.items()}

# Predict the sentiment
outputs = model(**inputs)
predictions = outputs.logits.argmax(dim=-1)

print(predictions)  # 1 for positive, 0 for negative


```
